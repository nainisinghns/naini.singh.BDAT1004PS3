```python
# Filter data for January and calculate the mean for each location..

january_data = data[data.index.month == 1]
january_means = january_data.mean()

print("January average wind speeds are:\n", january_means)
```

```python
# Downsample to yearly frequency

yearly = data.resample('Y').mean()

print("Yearly data is as:\n", yearly.head())
```

```python
# Downsampling to monthly frequency

monthly = data.resample('M').mean()

print("Monthly data is as:\n", monthly.head())
```

```python
# Downsample to weekly frequency

weekly = data.resample('W').mean()

print("Weekly data is as:\n", weekly.head())
```
