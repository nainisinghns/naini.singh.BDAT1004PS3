```python
import pandas as pd #Import pandas library
```

```python
import matplotlib.pyplot as plt #Import matplotlib library
```

```python
file_path = 'C:/Users/shaha/Desktop/BDAT/Problem Set 03/us-marriages-divorces-1867-2014.csv' #Locates the dataset on the local sytstem
```

```python
data = pd.read_csv(file_path)
```

```python
print(data.head()) #Displays first few rows of the dataset proving it is loaded correctly
```

```python
#Creating the line plot
```

```python
# This Plots the data
plt.figure(figsize=(12, 6))

# This Plots marriages per 1000 population
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per 1000', color='blue')

# This Plots divorces per 1000 population
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per 1000', color='red')

# We Add labels, title, and legend
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number per 1000 People', fontsize=12)
plt.title('Marriages and Divorces per Capita in the U.S. (1867–2014)', fontsize=14)
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()

# Displays the plot
plt.show()
```
